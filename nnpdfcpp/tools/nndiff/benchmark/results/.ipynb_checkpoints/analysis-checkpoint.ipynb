```python
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
```

```python
pdf= 'NNPDF30_nlo_as_0118'
fl = ['Gluon', 'Singlet','V','V3','V8','T3','T8']
```

```python
for ifl in fl:
    f = plt.figure()
    for irep in range(1,101):
        data = np.loadtxt(pdf + '_' + str(irep) + '_' + ifl+ '.dat')
        if irep == 1:
            plt.plot(data[:,0],data[:,1]/data[:,2], 'g', label='NNPDF/Analytic')
            plt.plot(data[:,0],data[:,1]/data[:,3], 'r', alpha=0.5, label='NNPDF/LHAPDF')
        else:
            plt.plot(data[:,0],data[:,1]/data[:,2], 'g')
            plt.plot(data[:,0],data[:,1]/data[:,3], 'r', alpha=0.5)
            
    plt.title(ifl + " - " + pdf)
    plt.legend(loc='best')
    plt.xscale('log')
    plt.xlabel('x')
    plt.xlim([1e-5,1])
    plt.ylim([0.7,1.3])
```

```python
replica = 2
for ifl in fl:
    
    f = plt.figure()
    plt.subplot(211)
    
    plt.title(ifl + " - " + pdf + " - Replica " + str(replica))
    data = np.loadtxt(pdf + '_' + str(replica) + '_' + ifl+ '.dat')
    plt.plot(data[:,0],data[:,1], 'g', label='NNPDF')  
    plt.legend(loc='best')
    plt.xscale('log')
    plt.xlim([1e-5,1])
    
    plt.subplot(212)
    plt.plot(data[:,0],data[:,4], 'r', label='NNPDFdev')            
    
    plt.legend(loc='best')
    plt.xscale('log')
    plt.xlabel('x')
    plt.xlim([1e-5,1])
    plt.ylim([-30,30])
```
